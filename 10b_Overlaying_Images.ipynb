```python
from PIL import Image, ImageDraw, ImageFont
import matplotlib.pyplot as plt
```

## *_Paste_*

```python
osprey_img = '/home/onkar/PyTorchForDLandCV/tools/PILImage/images/Osprey01.png'
egret_img = '/home/onkar/PyTorchForDLandCV/tools/PILImage/images/pythonlogo_small.png'

fig, axes = plt.subplots(1, 3, figsize=(30, 10))

with Image.open(osprey_img) as img1, Image.open(egret_img) as img2:   
    
    w1, h1 = img1.size
    print(f'Image 1 size : (W={w1}, H={h1})')
    print(f'Image 1 Mode : {img1.mode}')
    axes[0].set_title('Image #1')
    # axes[0].grid(True)
    axes[0].imshow(img1)
    
    w2, h2 = img2.size
    print(f'Image 2 size : (W={w2}, H={h2})')
    print(f'Image 2 Mode : {img2.mode}')
    axes[1].set_title('Image #2')
    # axes[1].grid(True)
    axes[1].imshow(img2)
        
    img1_copy = img1.copy()
    img1_copy.paste(img2, (w1//2 - w2//2, h1//2 - h2//2), img2) # paste the second image on the first image at the center, 3rd argument is the mask 
    
    axes[2].set_title('Blended Image')
    # axes[1].grid(True)
    axes[2].imshow(img1_copy)

    # show the plot
    plt.show()
```
