```python
from PIL import Image, ImageChops
import matplotlib.pyplot as plt
```

## *_alpha_composite_*

```python
osprey_img = '/home/onkar/PyTorchForDLandCV/tools/PILImage/images/Osprey01.png'
mask_img = '/home/onkar/PyTorchForDLandCV/tools/PILImage/images/pythonlogo.png'


## Adds two images, dividing the result by scale and adding the offset. If omitted, scale defaults to 1.0, and offset to 0.0. https://pillow.readthedocs.io/en/stable/reference/ImageChops.html#PIL.ImageChops.add
with Image.open(osprey_img) as img1, Image.open(mask_img) as mask:
    fig, axes = plt.subplots(1, 3, figsize=(30, 10))
    
    w1, h1 = img1.size
    print(f'Image 1 size : (W={w1}, H={h1})')
    print(f'Image 1 Mode : {img1.mode}')  
    axes[0].set_title('Image #1')
    axes[0].imshow(img1)
    
    w2, h2 = mask.size
    print(f'Image 2 size : (W={w2}, H={h2})')
    print(f'Image 2 Mode : {mask.mode}')
    axes[1].set_title('Mask image')
    axes[1].imshow(mask)
    
    # if img1.mode != img2.mode:
    #     img2 = img2.convert(img1.mode)
    
    # if img1.size != img2.size:
    #     img2 = img2.resize(img1.size)
    
    changed_img = Image.alpha_composite(img1.convert(mask.mode), mask)
    
    axes[2].set_title(f"Image after applying mask")
    axes[2].imshow(changed_img)
    
    plt.show()
```

## *_composite_*

```python
osprey_img = '/home/onkar/PyTorchForDLandCV/tools/PILImage/images/Osprey01.png'
stork_img = '/home/onkar/PyTorchForDLandCV/tools/PILImage/images/stork.png'
mask_img = '/home/onkar/PyTorchForDLandCV/tools/PILImage/images/pythonlogo.png'

with Image.open(osprey_img) as img1, Image.open(stork_img) as img2, Image.open(mask_img) as mask:
    fig, axes = plt.subplots(1, 4, figsize=(40, 10))
    
    w1, h1 = img1.size
    print(f'Image 1 size : (W={w1}, H={h1})')
    print(f'Image 1 Mode : {img1.mode}')  
    axes[0].set_title('Image #1')
    axes[0].imshow(img1)
    
    w2, h2 = img2.size
    print(f'Image 2 size : (W={w2}, H={h2})')
    print(f'Image 2 Mode : {img2.mode}')  
    axes[1].set_title('Image #2')
    axes[1].imshow(img2)
    
    w3, h3 = mask.size
    print(f'Mask Image size : (W={w3}, H={h3})')
    print(f'Mask Image Mode : {mask.mode}')
    axes[2].set_title('Mask image')
    axes[2].imshow(mask)
    
    if img1.mode != img2.mode:
        img2 = img2.convert(img1.mode)
    
    if img1.size != img2.size:
        img2 = img2.resize(img1.size)
    
    changed_img = Image.composite(img1, img2, mask.convert('L')) # If we change this to mask only, then we get differnt image
    
    axes[3].set_title(f"Image after applying mask")
    axes[3].imshow(changed_img)
    
    plt.show()
```
