```python
from PIL import Image, ImageDraw, ImageFont
import matplotlib.pyplot as plt
```

## *_Blend_*

```python
osprey_img = '/home/onkar/PyTorchForDLandCV/tools/PILImage/images/Osprey01.png'
egret_img = '/home/onkar/PyTorchForDLandCV/tools/PILImage/images/pythonlogo.png'
# egret_img = '/home/onkar/PyTorchForDLandCV/tools/PILImage/images/stork.png'


fig, axes = plt.subplots(1, 3, figsize=(30, 10))

with Image.open(osprey_img) as img1, Image.open(egret_img) as img2:   
    
    w1, h1 = img1.size
    print(f'Image 1 size : (W={w1}, H={h1})')
    print(f'Image 1 Mode : {img1.mode}')
    axes[0].set_title('Image #1')
    # axes[0].grid(True)
    axes[0].imshow(img1)
    
    w2, h2 = img2.size
    print(f'Image 2 size : (W={w2}, H={h2})')
    print(f'Image 2 Mode : {img2.mode}')
    axes[1].set_title('Image #2')
    # axes[1].grid(True)
    axes[1].imshow(img2)
    
    # The Image.blend function requires both images to be the same size and mode in order to work properly.
    if img1.mode != img2.mode:
        img2 = img2.convert(img1.mode)
    
    if img1.size != img2.size:
        img2 = img2.resize(img1.size)
    
    blended_img = Image.blend(img1, img2, 0.5)
    
    axes[2].set_title('Blended Image')
    # axes[1].grid(True)
    axes[2].imshow(blended_img)

    # show the plot
    plt.show()
```

## *_Composite_*

```python
osprey_img = '/home/onkar/PyTorchForDLandCV/tools/PILImage/images/Osprey01.png'
egret_img = '/home/onkar/PyTorchForDLandCV/tools/PILImage/images/pythonlogo.png'
# egret_img = '/home/onkar/PyTorchForDLandCV/tools/PILImage/images/stork.png'


fig, axes = plt.subplots(1, 4, figsize=(40, 10))

with Image.open(osprey_img) as img1, Image.open(egret_img) as img2:   
    
    w1, h1 = img1.size
    print(f'Image 1 size : (W={w1}, H={h1})')
    print(f'Image 1 Mode : {img1.mode}')
    axes[0].set_title('Image #1')
    # axes[0].grid(True)
    axes[0].imshow(img1)
    
    w2, h2 = img2.size
    print(f'Image 2 size : (W={w2}, H={h2})')
    print(f'Image 2 Mode : {img2.mode}')
    axes[1].set_title('Image #2')
    # axes[1].grid(True)
    axes[1].imshow(img2)
    
    # The Image.blend function requires both images to be the same size and mode in order to work properly.
    if img1.mode != img2.mode:
        img2 = img2.convert(img1.mode)
    
    if img1.size != img2.size:
        img2 = img2.resize(img1.size)
    
    # Create a mask image on fly 
    mask_img = Image.new('L', img2.size, 127) # 127 - midway - combined image, 0 = Only Osprey, 255 = only Logo
    axes[2].set_title('Mask Image')
    # axes[1].grid(True)
    axes[2].imshow(mask_img)
    
    composite_img = Image.composite(img1, img2, mask_img)
    axes[3].set_title('Composite Image')
    # axes[1].grid(True)
    axes[3].imshow(composite_img)

    # show the plot
    plt.show()
```
