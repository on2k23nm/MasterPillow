```python
from PIL import Image, ImageDraw, ImageFont
import matplotlib.pyplot as plt
```

Link to Font file : https://www.fontspace.com/category/truetype



```python
bird_img = '/home/onkar/PyTorchForDLandCV/tools/PILImage/images/Osprey01.png'

fig, axes = plt.subplots(1, 2, figsize=(20, 10))

with Image.open(bird_img) as bimg:
    copy_bimg = bimg.copy()
    draw_bimg = ImageDraw.Draw(copy_bimg) # Create a draw object
    
    # Rectangle 
    # https://pillow.readthedocs.io/en/stable/reference/ImageDraw.html#PIL.ImageDraw.ImageDraw.rectangle
    
    draw_bimg.rectangle((450, 100, 850, 400), outline='black', width=2, fill='red') # left, top, right, bottom
    
    # Ellipse
    # https://pillow.readthedocs.io/en/stable/reference/ImageDraw.html#PIL.ImageDraw.ImageDraw.ellipse
    
    draw_bimg.ellipse((700, 600, 1200, 1000), outline='black', width=2, fill='green') # left, top, right, bottom
    
    # Polygon
    # https://pillow.readthedocs.io/en/stable/reference/ImageDraw.html#PIL.ImageDraw.ImageDraw.polygon
    
    W, H = copy_bimg.size
    draw_bimg.polygon(((0, 400), (750, 850), (580, H), (0, H)), outline='black', width=2, fill='cyan')
    
    # Text
    # https://pillow.readthedocs.io/en/stable/reference/ImageDraw.html#PIL.ImageDraw.ImageDraw.text
    # https://pillow.readthedocs.io/en/stable/reference/ImageFont.html#PIL.ImageFont.truetype
    # https://pillow.readthedocs.io/en/stable/handbook/text-anchors.html
    ## Downloaded Rotulonahand-aGyx.ttf from - https://www.fontspace.com/category/truetype
    
    font = ImageFont.truetype('Rotulonahand-aGyx.ttf', size=100)
    draw_bimg.text((1000, 200), "Osprey", font=font) # Position, text
    
    axes[0].set_title('Original Image')
    # axes[0].grid(True)
    axes[0].imshow(bimg)
    
    axes[1].set_title('Image with shape drawn')
    # axes[1].grid(True)
    axes[1].imshow(copy_bimg)

    # show the plot
    plt.show()
```
