```python
from PIL import Image, ImageOps
import matplotlib.pyplot as plt
```

```python
# Deform the image.
# 
# https://pillow.readthedocs.io/en/stable/reference/ImageOps.html#PIL.ImageOps.deform

bird_img = '/home/onkar/PyTorchForDLandCV/tools/PILImage/images/Osprey01.png'

fig, axes = plt.subplots(1, 2, figsize=(20, 10))

class Deformer():
    def getmesh(self, img):
        # define a shape in the original image
        w, h = bimg.size
        src_left = (0, 0, 0, h, w//2, h, w//2, 0) 
        src_right = (w//2, 0, w//2, h, w, h, w, 0)
        
        # define a rectangle in the new image
        target_right = (w//2, 0, w, h)   # left, top, right, bottom
        target_left = (0, 0, w//2, h)
        
        # return all shapes
        return [(target_right, src_left), (target_left, src_right)]

## -- Invert colors --
with Image.open(bird_img) as bimg:
    deformed_bimg = ImageOps.deform(bimg, Deformer())
    
    # axes[0].grid(True)
    axes[0].set_title('Original Image')
    axes[0].imshow(bimg)

    # axes[1].grid(True)
    axes[1].set_title('Deformed Image')
    axes[1].imshow(deformed_bimg)

    # show the plot
    plt.show()
```
