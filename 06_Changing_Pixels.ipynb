# _*Changing a single Pixel in the image*_

```python
# Convert image to 1-bit grayscale image
from PIL import Image, ImageFilter
import matplotlib.pyplot as plt
import numpy as np
import pprint
from matplotlib.patches import Rectangle
```

```python
bird_img = '/home/onkar/PyTorchForDLandCV/tools/PILImage/images/Osprey01_tiny.png'

fig, axes = plt.subplots(1, 2, figsize=(20, 10))

with Image.open(bird_img) as bimg:
    new_bimg = bimg.copy()
    new_bimg.putpixel((260, 260), (255, 0, 0)) # Put Red Pixel in the middle of the image
    
    axes[0].imshow(bimg)
    axes[0].set_title('Original Image')

    axes[1].imshow(new_bimg)
    axes[1].set_title('Image with Red Pixel at Center')
    
    # Annotate the coordinate on the new image
    axes[1].annotate('(260, 260)', # This is the text to display
                 (260, 260), # These are the coordinates to position the text
                 textcoords="offset points", # how to position the text
                 xytext=(0,10), # distance from text to points (x,y)
                 ha='center') # horizontal alignment can be left, right or center


    # Create a rectangle patch with the bottom left corner slightly offset from the pixel
    # The size (10, 10) can be adjusted as needed
    rect = Rectangle((260-5, 260-5), 10, 10, linewidth=1, edgecolor='b', facecolor='none')

    # Add the rectangle to the axes
    axes[1].add_patch(rect)

    # show the plot
    plt.show()
```

## _*Change large numbers of Pixels in an image*_

```python
# Note that below operation, changing Pixels with Pillow in 2 for loops is very slow operation - if you want to do this more regularly, you will have to use numpy arrays

bird_img = '/home/onkar/PyTorchForDLandCV/tools/PILImage/images/Osprey01_tiny.png'

fig, axes = plt.subplots(1, 2, figsize=(20, 10))

with Image.open(bird_img) as bimg:
    new_bimg = bimg.copy()
    # new_bimg.putpixel((260, 260), (255, 0, 0)) # Put Red Pixel in the middle of the image
    
    W, H = new_bimg.size
    for x in range(W):
        for y in range(H):
            colors = new_bimg.getpixel((x, y))
            if colors[0] > 200:
                new_bimg.putpixel((x, y), (255, 0, 0))
    
    axes[0].imshow(bimg)
    axes[0].set_title('Original Image')

    axes[1].imshow(new_bimg)
    axes[1].set_title('Image with Red Pixel at Center')

    # show the plot
    plt.show()
```
